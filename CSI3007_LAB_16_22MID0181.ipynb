```python
# Import necessary libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Load and display the original image ---
# Load the image and convert it to a NumPy array.
# The `with` statement handles closing the file automatically.
try:
    with Image.open('C:/Users/dhars/Pictures/Screenshots/large.jpeg') as img:
        im = np.asarray(img)
except FileNotFoundError:
    print("Error: The image file was not found. Please check the file path.")
    exit()

# Create a new figure and a single subplot for the original image
fig, ax = plt.subplots(figsize=(8, 6))
ax.imshow(im)
ax.set_title('Original Bug Image')
ax.axis('off')  # Hide the axes for a cleaner image display

# --- 2. Create plots for individual color channels ---
# This visualizes the red, green, and blue components of the image.
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15, 5))

# Red channel
red_channel = im[:, :, 0]
ax1.imshow(red_channel, cmap='Reds_r')
ax1.set_title('Red Channel')
ax1.axis('off')

# Green channel
green_channel = im[:, :, 1]
ax2.imshow(green_channel, cmap='Greens_r')
ax2.set_title('Green Channel')
ax2.axis('off')

# Blue channel
blue_channel = im[:, :, 2]
ax3.imshow(blue_channel, cmap='Blues_r')
ax3.set_title('Blue Channel')
ax3.axis('off')

fig.suptitle('Individual Color Channels', fontsize=16)


# --- 3. Create an image with an alternative color map ---
# Change the color map to emphasize different features.
fig, ax = plt.subplots(figsize=(8, 6))
# Display the red channel with a 'hot' color map
ax.imshow(red_channel, cmap='hot')
ax.set_title('Image with "Hot" Color Map')
ax.axis('off')

# --- 4. Create a histogram of pixel intensity ---
# A histogram is a great way to visualize the distribution of colors.
fig, ax = plt.subplots(figsize=(8, 6))
# ravel() flattens the multi-dimensional array into a 1D array
ax.hist(im.ravel(), bins=256, color='gray', alpha=0.7)
ax.set_title('Pixel Intensity Histogram')
ax.set_xlabel('Pixel Intensity')
ax.set_ylabel('Frequency')
ax.grid(True)

# --- 5. Display all plots ---
# This is the command you were missing. It displays all the figures created.
plt.tight_layout() # Adjusts subplot params for a tight layout
plt.show()
```

# with Black & White u will get the correct distribution of pixels

```python
# Import necessary libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Define the image path (no space at the beginning!)
image_path = r"C:\Users\VARSHITHA V\Downloads\IMAGE-1.avif"

# --- 1. Open the color image ---
try:
    # Open the original color image
    img_color = Image.open(image_path)
except FileNotFoundError:
    print("Error: The image file was not found. Please check the file path.")
    exit()

# --- 2. Convert to a black and white (grayscale) image ---
# The 'L' mode converts the image to grayscale, using one channel (luminance).
img_grayscale = img_color.convert('L')
im_grayscale_array = np.asarray(img_grayscale)

# --- 3. Display the grayscale image and its histogram ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Display the grayscale image
ax1.imshow(im_grayscale_array, cmap='gray')
ax1.set_title('Grayscale Bug Image')
ax1.axis('off')  # Hide axes for cleaner image display

# Plot the histogram for the grayscale image
ax2.hist(im_grayscale_array.ravel(), bins=256, range=[0, 256], color='k', alpha=0.7)
ax2.set_title('Grayscale Pixel Intensity Histogram')
ax2.set_xlabel('Pixel Intensity (0=Black, 255=White)')
ax2.set_ylabel('Frequency (Number of Pixels)')
ax2.grid(True)

plt.tight_layout()  # Adjusts subplot params for a tight layout
plt.show()
```
