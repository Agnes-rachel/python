# read a csv file and create it as dataframe

```python
import pandas as pd
df = pd.read_csv('products-100.csv')
df.head()
```

# APPEND TO CSV

```python
new_row={'Index':100,'Name':'lamp','Description':'bright','Brand':'ind','Category':'gadget','Price':464,'Currency':'USD','Stock':823,'EAN':253727724873,'Color':'black','Size':'small','Availability':'in_stock','Internal ID':64}
```

```python
new_data_df=pd.DataFrame([new_row])
```

```python
df = pd.concat([df, new_data_df], ignore_index=True)#concating new data frame with existing dataframe
```

```python
df.to_csv('products-100.csv', index=False)#new dataframe copied to csv
```

# read a csv chunk by chunk

```python
from io import StringIO
import pandas as pd
df_chunk = pd.read_csv('products-100.csv', usecols=["Name", "Brand"])
print(df_chunk.head())
```

```python
import pandas as pd

# Read CSV file in chunks of 10 rows
chunk_iter = pd.read_csv('products-100.csv', chunksize=10)

# Iterate through each chunk
for i, chunk in enumerate(chunk_iter):
    print(f"\n--- Chunk {i+1} ---")
    print(chunk)
```

# write a numeric data into a csv file

```python
import pandas as pd

# Create numeric data
data = {
    "ID": [1, 2, 3, 4, 5],
    "Value1": [10, 20, 30, 40, 50],
    "Value2": [5.5, 6.7, 7.8, 8.9, 9.1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Write to CSV file
df.to_csv("numeric_data.csv", index=False)

print("CSV file written successfully!")
```

# write text data into a csv file

```python
import pandas as pd

# Create text data
data = {
    "Name": ["Agnes", "jayanth", "latha", "edwin", "radha"],
    "City": ["chennai", "banglore", "chennai", "america", "vellore"],
    "Profession": ["Doctor", "Engineer", "Teacher", "Artist", "Designer"]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Write to CSV file
df.to_csv("text_data.csv", index=False)

print("Text CSV file written successfully!")
```
