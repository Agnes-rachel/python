## PDF Text Extraction using PyMuPDF

This example demonstrates how to use the PyMuPDF (fitz) library to open a PDF file, extract text from each page, and display it. We'll create a document instance, loop through the pages, and print the text content with comments explaining each step.


```python
# Step 1: Import the fitz module (PyMuPDF)
import fitz  # PyMuPDF library

# Step 2: Open the PDF file using fitz.open()
# Argument: Provide the path to the PDF file
doc = fitz.open("D:/Downloads/pdf.pdf")  # This is the instance of the PDF document

# Step 3: Loop through each page in the PDF
for page_num, page in enumerate(doc):
    # Extract text from the page
    text = page.get_text()  # Extracts the text as a string
    
    # Print page number and its content
    print(f"--- Page {page_num + 1} ---")
    print(text)
    print("\n" + "="*40 + "\n")

# Step 4: Close the document after use
doc.close()
```
