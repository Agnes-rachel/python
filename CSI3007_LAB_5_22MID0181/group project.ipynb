```python
# Banking system with nested dictionary & list
bank_data = {
    'C001': {
        'name': 'Alice',
        'age': 30,
        'transactions': [5000, -1500, 2200],
    },
    'C002': {
        'name': 'Bob',
        'age': 25,
        'transactions': [7000, -2000, 300],
    },
    'C003': {
        'name': 'Charlie',
        'age': 35,
        'transactions': [10000, -500, -4000],
    },
}

# Function to add new customer
def add_customer():
    cid = input("Enter new customer ID (e.g., C004): ")
    name = input("Enter customer name: ")
    age = int(input("Enter customer age: "))
    transactions = list(map(int, input("Enter initial transactions (space separated): ").split()))
    bank_data[cid] = {
        'name': name,
        'age': age,
        'transactions': transactions
    }
    print(f" Customer {cid} added.\n")

# Function to view specific customer using .get()
def view_customer():
    cid = input("Enter customer ID to view: ")
    data = bank_data.get(cid)# does not throw error if key not found
    if data:
        name, age, transactions = data.get('name', 'N/A'), data.get('age', 'N/A'), data.get('transactions', [])
        print(f"\nCustomer ID: {cid}")
        print(f"Name: {name}")
        print(f"Age: {age}")
        print("Transaction History:")
        
        # Using enumerate to list transactions with index
        for i, txn in enumerate(transactions, 1):
            print(f"  {i}. ₹{txn}")

        #  Using any to check for any negative transaction
        if any(t < 0 for t in transactions):
            print(" Withdrawals detected.")
        else:
            print(" All transactions are deposits.")
    else:
        print("Customer not found.\n")

# Function to print last 5 added customers using zip and sorted
def show_last_five_entries():
    print("\n Last 5 entries in the banking system:")
    
    # Using sorted and lambda to sort by name
    for cid, info in sorted(bank_data.items(), key=lambda x: x[1]['name']):
        print(f"ID: {cid}, Name: {info['name']}, Age: {info['age']}")

# Looping through customers and showing balances
print(" All Customers in the System:")
for cid, info in bank_data.items():
    name, age, transactions = info['name'], info['age'], info['transactions']
    print(f"ID: {cid}, Name: {name}, Age: {age}, Total Balance: ₹{sum(transactions)}")

# Adding two new customers
for _ in range(2):
    add_customer()

# View a specific customer
view_customer()

# Show last 5 entries sorted by name
show_last_five_entries()

# Pair customer names with balances using zip
print("\n Customer Names & Balances (via zip):")
names = [data['name'] for data in bank_data.values()]
balances = [sum(data['transactions']) for data in bank_data.values()]
for name, balance in zip(names, balances):
    print(f"{name}: ₹{balance}")
```
